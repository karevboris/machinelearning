### Описание датасета и задачи

Данные представляют собой статистику о сейсмической активности в шахтах. Задача классификации - определение класса сейсмической опасности для следующей рабочей смены, основываясь на характеристиках активностей в прошлые смены. Энергия измеряется в Джоулях.

Записей: 2584

Аттрибуты: 18 + class

Классификация на:

    -"hazardous state" (опасное состояние, class 1)    :  170  (6.6%) (если class=1 в таблице, то это означает, что сейсмический удар высокой энергии произошел в следующей смене)
    
    -"non-hazardous state" (не опасное состояние, class 0): 2414 (93.4%) (означает, что в следующей смене не возникло сильных сейсмических колебаний)

Пропущенные значения аттрибутов отсутствуют.

Описание аттрибутов:

    -seismic: результат оценки сейсмической опасности во время рабочей смены в шахте, полученный сейсмическим методом (a - отсутствие опасности, b - низкая опасность, c - высокая опасность, d - состояние опасности);
 
    -seismoacoustic: результат оценки сейсмической опасности во время рабочей смены в шахте, полученный сейсмоакустическим методом;
 
    -shift: информация о типе смены (W - добыча угля, N - подготовительная);
    
    -genergy: сейсмическая энергия, зарегистрированная в предыдущей смене наиболее активным геофоном (GMax) из геофонов, контролирующих подземную добычу угля;
 
    -gpuls: количество импульсов, записанных GMax в предыдущей смене;
    
    -gdenergy: отклонение энергии, зарегистрированное GMax в течение предыдущей смены, от средней энергии, зарегистрированной в течение восьми предыдущих смен;
 
    -gdpuls: отклонение числа импульсов, записанных в течение предыдущей смены на GMax, от среднего числа импульсов, зарегистрированных в течение восьми предыдущих смен;
  
    -ghazard: результат оценки сейсмической опасности во время рабочей смены в шахте, полученной сейсмоакустическим методом, основанный только на показаниях от GMax;

    -nbumps: количество сейсмических ударов, зарегистрированных в предыдущей смене;

    -nbumps2: количество сейсмических ударов (в диапазоне энергии (10^2,10^3)) зарегистрированных в предыдущей смене;

    -nbumps3: количество сейсмических ударов (в диапазоне энергии (10^3,10^4)) зарегистрированных в предыдущей смене;

    -nbumps4: количество сейсмических ударов (в диапазоне энергии (10^4,10^5)) зарегистрированных в предыдущей смене;
    
    -nbumps5: количество сейсмических ударов (в диапазоне энергии (10^5,10^6)) зарегистрированных в предыдущей смене;
 
    -nbumps6: количество сейсмических ударов (в диапазоне энергии (10^6,10^7)) зарегистрированных в предыдущей смене;
 
    -nbumps7: количество сейсмических ударов (в диапазоне энергии (10^7,10^8)) зарегистрированных в предыдущей смене;
 
    -nbumps89: количество сейсмических ударов (в диапазоне энергии (10^8,10^10)) зарегистрированных в предыдущей смене;
 
    -energy: суммарная энергия сейсмических ударов, зарегистрированных в течение предыдущей смены;
 
    -maxenergy: максимальная энергия сейсмических ударов, зарегистрированных в течение предыдущей смены.

# Подготовка данных

Разбиение на количесвенные и категориальные признаки

```python
categorical_columns = [c for c in data.columns if data[c].dtype.name == 'object']
numerical_columns   = [c for c in data.drop(('class'), axis = 1).columns if data[c].dtype.name != 'object']
print(categorical_columns)
print(numerical_columns)
```

Убеждаемся что пропущенные значения аттрибутов отсутствуют.

#### Векторизация категориальных признаков

```python
data_describe = data.describe(include = [object])
binary_columns    = [c for c in categorical_columns if data_describe[c]['unique'] == 2]
nonbinary_columns = [c for c in categorical_columns if data_describe[c]['unique'] > 2]
print(binary_columns, nonbinary_columns)
```

```python
data['seismic'].unique()
```

```python
data['shift'].unique()
```

```python
data.at[data['seismic'] == 'a', 'seismic'] = 0
data.at[data['seismic'] == 'b', 'seismic'] = 1
data['seismic'].describe()
```

```python
data.at[data['shift'] == 'N', 'shift'] = 0
data.at[data['shift'] == 'W', 'shift'] = 1
data['shift'].describe()
```

К небинарными признакам применим метод _векторизации_, 
который заключается в следующем.

Признак `j`, принимающий `s` значений, заменим на `s` признаков, принимащих значения `0` или `1`,
в зависимости от того, чему равно значение исходного признака `j`.

```python
data_nonbinary = pd.get_dummies(data[nonbinary_columns])
print(data_nonbinary.columns)
```

#### Нормализация количественных признаков

Многие алгоритмы машинного обучения чувствительны к масштабированию данных.
К таким алгоритмам, например, относится метод ближайших соседей, машина опорных векторов и др.

В этом случае количественные признаки полезно _нормализовать_.
Это можно делать разными способами.
Например, каждый количественный признак приведем к нулевому среднему и единичному среднеквадратичному отклонению:

```python
data_numerical = data[numerical_columns]
data_numerical.describe()
```

```python
data_numerical = (data_numerical - data_numerical.mean(axis = 0))/data_numerical.std(axis = 0)
```

```python
data_numerical.describe()
```

#### Соединяем все в одну таблицу

```python
data = pd.concat((data_numerical, data_nonbinary, data[binary_columns], data['class']), axis = 1)
print(data.shape)
```

```python
data.describe()
```

### Обучение на данных полносвязной (неглубокой) нейронной сети

#### Обучаем классификатор

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp_model = MLPClassifier(hidden_layer_sizes = (100,), solver = 'lbfgs', 
                          activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)
```

```python
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
```

```python
print(np.mean(y_train != y_train_pred), np.mean(y_test != y_test_pred))
```

```python
print(mlp_model.score(X_train, y_train), mlp_model.score(X_test, y_test))
```

```python

print(1 - mlp_model.score(X_test, y_test))
```

#### Зависимость ошибки от параметра alpha

```python
alpha_arr = np.logspace(-3, 2, 21)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (100,), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'test')
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'train')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

```python
alpha_opt = alpha_arr[test_err == min_test_err]
print(alpha_opt)
```

```python
alpha_opt = alpha_opt[0]
```

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (35,35),
                          solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)

y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print(np.mean(y_train != y_train_pred))
print(np.mean(y_test != y_test_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test_pred, y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_pred))
```

По сравнению с методом ближайших соседей, нейронная сеть уже выдаёт меньшее количество ошибок второго рода, стало больше ошибок первого рода, однако и число правильно определённых объектов первого класса стало больше.

#### Варирование количества скрытых слоёв и нейронов

```python
classifiers = [
    MLPClassifier(hidden_layer_sizes = (20,), solver = 'lbfgs', activation = 'logistic', random_state = 42),
    MLPClassifier(hidden_layer_sizes = (50,), solver = 'lbfgs', activation = 'logistic', random_state = 42),
    MLPClassifier(hidden_layer_sizes = (20,50), solver = 'lbfgs', activation = 'logistic', random_state = 42),
    MLPClassifier(hidden_layer_sizes = (50, 50), solver = 'lbfgs', activation = 'logistic', random_state = 42),
    MLPClassifier(hidden_layer_sizes = (100,100), solver = 'lbfgs', activation = 'logistic', random_state = 42),
]
```

```python
for classifier in classifiers:
    classifier.fit(X_train, y_train)
    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    alpha_arr = np.logspace(-3, 3, 21)
    test_err = []
    train_err = []
    for alpha in alpha_arr:
        mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = classifier.hidden_layer_sizes, solver = 'lbfgs', activation = 'logistic', random_state = 42)
        mlp_model.fit(X_train, y_train)

        y_train_pred = mlp_model.predict(X_train)
        y_test_pred = mlp_model.predict(X_test)
        train_err.append(np.mean(y_train != y_train_pred))
        test_err.append(np.mean(y_test != y_test_pred))
    
    plt.semilogx(alpha_arr, train_err, 'b-o')
    plt.semilogx(alpha_arr, test_err, 'r-o')
    plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
    plt.title('Error vs. alpha')
    plt.xlabel('alpha')
    plt.ylabel('error')
    pass
```

При использовании разного количества скрытых слоёв и нейронов, оптимальный параметр alpha не менятеся.

```python
classifier = MLPClassifier(hidden_layer_sizes = (35, 35), solver = 'lbfgs', activation = 'logistic', random_state = 42)
classifier.fit(X_train, y_train)
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
alpha_arr = np.logspace(-3, 2, 21)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = classifier.hidden_layer_sizes, solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'test')
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'train')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

```python
num_arr = np.logspace(0, 3, 21)
test_err = []
train_err = []
for num in num_arr:
    mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (int(num), int(num)), solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    
plt.semilogx(num_arr, train_err, 'b-o', label = 'test')
plt.semilogx(num_arr, test_err, 'r-o', label = 'train')
plt.xlim([np.max(num_arr), np.min(num_arr)])
plt.title('Error vs. numbers of neirons')
plt.xlabel('number')
plt.ylabel('error')
plt.legend()
pass
```
